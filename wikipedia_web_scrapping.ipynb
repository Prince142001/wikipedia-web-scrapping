```python
from bs4 import BeautifulSoup
import requests

url = "https://www.wikipedia.org/"

# response = requests.get(url)
# response

header={
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"
}
response= requests.get(url, headers=header)
response
```

```python
import pandas as pd

html_content=BeautifulSoup(response.content, "html.parser")
# html_content

other_project_title_data  = []
other_project_title   = html_content.select("span.other-project-title")

for i in other_project_title:
  other_project_title_data.append(i.text.strip())

# other_project_title_data

other_project_tagline_data = []
other_project_tagline= html_content.select("span.other-project-tagline")

for i in other_project_tagline:
  other_project_tagline_data.append(i.text.strip())

other_project_link_data=[]
other_project_link=html_content.select("a.other-project-link")

for i in other_project_link:
  link=i.get('href')
  if link:
    other_project_link_data.append(link)

df=pd.DataFrame({
    "Title": other_project_title_data,
    "Tag" : other_project_tagline_data,
    "Link" : other_project_link_data
})

df
```

```python
import pandas as pd
import re

html_content=BeautifulSoup(response.content, "html.parser")
# html_content

other_project_title_data  = []
other_project_title   = html_content.select("span.other-project-title")

for i in other_project_title:
  other_project_title_data.append(i.text.strip())

# other_project_title_data

other_project_tagline_data = []
other_project_tagline= html_content.select("span.other-project-tagline")

for i in other_project_tagline:
  other_project_tagline_data.append(i.text.strip())

other_project_link_data=[]
other_project_link=html_content.select("a.other-project-link")

for i in other_project_link:
  link=i.get('href')
  if link:
    clean_link = re.sub("/", "", link)
    other_project_link_data.append(clean_link)

df=pd.DataFrame({
    "Title": other_project_title_data,
    "Tag" : other_project_tagline_data,
    "Link" : other_project_link_data
})

df
```

### URL Join

```python
from urllib.parse import urljoin

import pandas as pd
import re

html_content=BeautifulSoup(response.content, "html.parser")
# html_content

other_project_title_data  = []
other_project_title   = html_content.select("span.other-project-title")

for i in other_project_title:
  other_project_title_data.append(i.text.strip())

# other_project_title_data

other_project_tagline_data = []
other_project_tagline= html_content.select("span.other-project-tagline")

for i in other_project_tagline:
  other_project_tagline_data.append(i.text.strip())

other_project_link_data=[]
other_project_link=html_content.select("a.other-project-link")

for i in other_project_link:
  link=i.get('href')
  if link:
    clean_link = re.sub("/", "", link)
    clickable_link=urljoin(url, clean_link)
    other_project_link_data.append(clickable_link)

df=pd.DataFrame({
    "Title": other_project_title_data,
    "Tag" : other_project_tagline_data,
    "Link" : other_project_link_data
})

df
```
